```python
import sympy as sp
```

```python
t, t0, d, L, k, K, a = sp.symbols(r"t t_0 d L k \lambda a")
```

```python
h = L*(t+a)**k/((t+a)**k+K**k)+d
h
```

```python
dh = h.diff(t)
ddh = dh.diff(t)
tmax = sp.solve(ddh, t)[0]
display(dh)
display(ddh)
display(tmax)
```

```python
vmax = (dh.subs(t, tmax).simplify().simplify())
vmax
```
