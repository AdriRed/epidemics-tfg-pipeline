```python
import networkx as nx
import numpy as np
from scipy import stats
import powerlaw  # Necesitas instalarlo: pip install powerlaw

def estimate_gamma_powerlaw(G, kmin=None):
    """
    Estima el exponente gamma de la ley de potencia para la distribución de grados
    
    Args:
        G: grafo de NetworkX
        kmin: grado mínimo a partir del cual ajustar (None = auto-detectar)
    
    Returns:
        gamma: exponente estimado
        kmin: grado mínimo usado
    """
    # Obtener grados de todos los nodos
    degrees = [d for n, d in G.degree()]
    degrees = np.array(degrees)
    
    # Filtrar grados > 0
    degrees = degrees[degrees > 0]
    
    if len(degrees) < 10:
        print("Advertencia: pocos nodos con grado > 0")
        return None, None
    
    # Usar la librería powerlaw para estimación robusta
    fit = powerlaw.Fit(degrees, xmin=kmin, discrete=True)
    
    return fit.alpha, fit.xmin

# Ejemplo: Crear una red scale-free con Barabási-Albert
G = nx.read_edgelist("./epidemics-tfg/nets/inf-openflights.edges", comments='%')

gamma, kmin = estimate_gamma_powerlaw(G)
print(f"Gamma estimado: {gamma:.3f}")
print(f"k_min: {kmin}")
```
